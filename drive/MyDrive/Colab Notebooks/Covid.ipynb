```python
import pandas as pd
```

```python
covid_data = pd.read_excel("covid_worldwide.xlsx")
covid_data.fillna(0, inplace=True)
covid_data.info()
```

```python
cases_comparision =  covid_data.loc[ 0:9, ["Country","Total Cases", "Total Deaths", "Total Recovered", "Active Cases"]]
cases_comparision.plot.bar(x="Country", rot=65, title="Covid case comparisions ")
```

```python
death_recovery_cases =  covid_data.loc[ 0:9, ["Country","Total Recovered","Total Deaths", ]]
death_recovery_cases.plot.barh(x="Country", stacked=True, title=" Comparision of death cases and recovery cases")
```

```python
affected_rate = covid_data.loc[ 0:9, ["Country","Total Cases"]]
affected_rate.plot.pie(y="Total Cases", labels=affected_rate["Country"], figsize=(10, 10), autopct="%1.1f%%", title="Covid impact rate in top 10 countries")
                                      
```
